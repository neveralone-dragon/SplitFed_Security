### 2.1.2 计算准确率的四个指标

```python
import numpy as np

def calculate_metrics(acc_test_list, convergence_threshold=0.01, window_size=5):
    avg_acc_test = np.mean(acc_test_list)
    max_acc_test = np.max(acc_test_list)
    min_acc_test = np.min(acc_test_list)

    # 计算收敛速度
    convergence_index = -1
    for i in range(len(acc_test_list) - window_size):
        acc_test_window = acc_test_list[i: i + window_size]
        if np.std(acc_test_window) < convergence_threshold:
            convergence_index = i
            break

    return round(avg_acc_test, 2), round(max_acc_test, 2), round(min_acc_test, 2), convergence_index

```
